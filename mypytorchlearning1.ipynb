one of the most common error in deep learning.: shape errors
Matrix multipliation operations works when Tensor _B is transposed.

```python
#shapes for matrix multiplication
import torch
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])

#trch.mm(tensor_A, tensor_B) #torch.mm is the same as torch.matmul
torch.matmul(tensor_A, tensor_B)
```

```python
tensor_A
```

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
#the matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

RESHAPING STACKING SQUEEZING  UNSQUEEZING TENSORS
* Reshaping - reshapes an input tensor to a defined shapes
* view - return a view of an input of a certain shape but keep the same memory as the original one
* stacking - combining multiple tensor ontop of each other. torch.vstark, torch.hstark. Torch.stack
* squeeze -m removeall `1`dimensions from a tensor.
* unsqueeze _ add 1 dimension target tensor
* permute- return a view point  with the dimensions permuted (swapped in a certain nway

```python
#create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
#change the view
z = x.view(1, 9)
z, z.shape
```

view shares the same memory as the original imput

```python
# change the view
z = x.view(1, 9)
z, z.shape
```

```python
#changes z changes x blc they both share the same memory
z[:, 0] = 5
z, x
```

```python
# stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

SQUEEZE AND UNSQUEEZE AND PERMUTATION

```python
# torch.squeeze() - removes all single dimensions frm a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

#remove extra dimension from x shaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
#torch.unsqueze() -nadd a single dimension to a target tensor at a specific dimension
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

#add extra dimension from x
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
#torch,permute = rearranges the dimensions of a target tensor in the specific order
x_original = torch.rand(size=(224, 224, 3)) # height, width, colour channels

#permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # colour channels, height, width
```

x_original shares the same memory as x_permuted

```python
x_original[0, 0, 0] = 728218
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

PYTORCH REPRODUCIBILITY
(trying to take random out of random)
How a Neural Network Learns:
- start with random numbers -> tensor operations ->updata random numbers to try and make them better representation of the data again-> again-> again
-to reduce the randomness in neural networks  and pytocrch comes the concept of random seed
essential what the random see does is "flavor' the randomness
more resources https://pytorch.org/docs/stable/notes/randomness.html


```python
torch.rand(3, 3)
```

```python
import torch
#create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
#lets make some random but reproducible tensors
import torch

#set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)
torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
